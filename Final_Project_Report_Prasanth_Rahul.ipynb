### PPV v/s Sensitivity Plot

The precision-sensitivity trade-off of a binary classification model across various decision thresholds is shown by a PPV (Positive Predictive Value) vs. Sensitivity plot, also known as a Precision-Recall curve. It demonstrates how accurately the model detects positive cases (sensitivity) while upholding a high level of precision. The curve demonstrates the model's capacity to choose a suitable threshold for particular use cases while balancing accurately recognizing genuine positives and avoiding false positives.

```python
# obtain the precision, recall and thresholds
precision, recall, thresholds = precision_recall_curve(y_test, predictions)

# calculating PPV for each threshold
ppv = []
for i in range(len(thresholds)):
    y_pred = (predictions >= thresholds[i]).astype(int)
    tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
    ppv.append(tp / (tp + fp))

# plot the PPV against sensitivity
plt.plot(recall[:-1], ppv)
plt.xlabel('Sensitivity')
plt.ylabel('PPV')
plt.title('PPV vs Sensitivity (Precision-Recall Curve)')
plt.show()
```

The model has a high rate of falsely identifying positive cases. It means that as it identifies more true positive cases correctly, it also falsely identifies more cases as positive.

```python
# function to calculate the metrics and their confidence intervals for a given threshold
def get_metrics(y_true, y_prob, threshold):
    y_pred = (y_prob >= threshold).astype(int)
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
    tpr = tp / (tp + fn) if (tp + fn) > 0 else 0 # sensitivity or recall
    tnr = tn / (tn + fp) if (tn + fp) > 0 else 0 # specificity
    ppv = tp / (tp + fp) if (tp + fp) > 0 else 0 # positive predictive value or precision
    npv = tn / (tn + fn) if (tn + fn) > 0 else 0 # negative predictive value
    fpr = fp / (fp + tn) if (fp + tn) > 0 else 0 # false positive rate
    fnr = fn / (fn + tp) if (fn + tp) > 0 else 0 # false negative rate
    fdr = fp / (fp + tp) if (fp + tp) > 0 else 0 # false discovery rate
    accuracy = (tp + tn) / (tp + tn + fp + fn)
    return tpr, tnr, ppv, npv, fpr, fnr, fdr, accuracy

#  function to calculate the metrics and their confidence intervals for all possible thresholds using the bootstrap percentile method
def get_metrics_ci(y_true, y_prob, num_reps=1000, ci=95):
    thresholds = np.unique(y_prob)[::-1]
    num_thresholds = len(thresholds)
    metrics = np.zeros((num_thresholds, 8))
    for i, threshold in enumerate(thresholds):
        metrics[i, :] = get_metrics(y_true, y_prob, threshold)
    ci_lower = (100 - ci) / 2
    ci_upper = ci + ci_lower
    metric_cis = []
    for i in range(8):
        metric_samples = resample(metrics[:, i], replace=True, n_samples=num_reps)
        ci = np.percentile(metric_samples, [ci_lower, ci_upper])
        metric_cis.append(ci)
    return metrics, metric_cis
```

```python
metrics, metric_cis = get_metrics_ci(y_test, predictions)

metric_names = ['Sensitivity', 'Specificity', 'Precision', 'Negative Predictive Value', 'False Positive Rate', 'False Negative Rate', 'False Discovery Rate', 'Accuracy']

for i, metric_name in enumerate(metric_names):
    print(f"{metric_name}: {metrics[-1, i]:.3f} ({metric_cis[i][0]:.3f}, {metric_cis[i][1]:.3f})")
```

### **Oversampling of the dataset**

For oversampling, we employ SMOTE (Synthetic Minority Oversampling Technique). By interpolating between the minority class instances that already exist, the SMOTE approach produces new instances of the minority class.

```python
from imblearn.over_sampling import SMOTE

#Splitting data into features and target variable
X = df.drop('corona_result', axis = 1)
y = df['corona_result']

X.drop('test_date', axis = 1, inplace = True)

# Over sampling the dataset
smote_sample = SMOTE()
X_over, y_over = smote_sample.fit_resample(X, y)

X_over.shape
```

### MLP (Multi-Layer Perceptron)

An artificial neural network known as an MLP (Multi-Layer Perceptron) is made up of several interconnected layers of nodes or neurons, comprising an input layer, one or more hidden layers, and an output layer. It is a potent supervised learning method used for many different applications, including regression and classification. Each layer's neurons use a nonlinear activation function to activate their weighted inputs, enabling the network to recognize intricate patterns in the data. Backpropagation training allows the model to understand complicated patterns and produce precise predictions by modifying weights to reduce the discrepancy between projected and actual outputs.

```python
#split into training and testing sets using oversampled data, with 80% for training and 20% for testing
X_train, X_test, y_train, y_test = train_test_split(X_over, y_over, test_size=0.2, random_state=42)
```

```python
from sklearn.neural_network import MLPClassifier

# Create MLPClassifier object and fit to training data
mlp = MLPClassifier(hidden_layer_sizes=(10, 10), max_iter=1000, random_state=42)
mlp.fit(X_train, y_train)
```

The Multi-Layer Perceptron (MLP) neural network model for classification tasks is represented by the MLPClassifier. It has two buried layers with a total of 10 neurons each. The random number generator seed is fixed to 42 by the random_state option, which also defines the maximum number of training iterations. In order to make precise classifications, this model seeks to discover intricate patterns in the data.

```python
# Use the trained model to make predictions on the test data
y_pred = mlp.predict(X_test)

# Evaluate the performance of the model
score = mlp.score(X_test, y_test)
print(f"Accuracy: {score}")
```

The MLP Classifier's accuracy score of 0.8618 means that for about 86.18% of the dataset's cases, the model correctly predicts the class labels. This shows that the model has developed the ability to recognize significant patterns and relationships in the data, leading to a relatively high level of overall prediction accuracy.
